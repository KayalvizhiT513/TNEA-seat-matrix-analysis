```python
import camelot
import pandas as pd
```

```python
def pdf_to_excel(pdf_path, excel_path):
    # Extract tables from each page of the PDF using camelot
    tables = camelot.read_pdf(pdf_path, pages='all', flavor='stream')

    # List to store all the tables from the PDF
    all_tables = []

    # Loop through each extracted table and append to the all_tables list
    for table in tables:
        all_tables.append(table.df)

    # Concatenate all tables into a single DataFrame
    full_table = pd.concat(all_tables, ignore_index=True)

    # Save the concatenated table to an Excel file
    full_table.to_excel(excel_path, index=False, engine='openpyxl')

    print(f"PDF converted and saved as {excel_path}")

# Example usage
pdf_path = '/content/Rank List 2024.xlsx.pdf'  # Specify the PDF file path
excel_path = 'Rank List 2024.xlsx'  # Specify the desired output Excel file path
pdf_to_excel(pdf_path, excel_path)
```

```python
allotment_2024_609 = pd.merge(prov_alot_r3_2024, ranklist_2024_sc, on='APPLN NO', how='inner')
# List columns with _x and _y suffixes (i.e., likely duplicates)
dup_cols = [col[:-2] for col in allotment_2024_609.columns if col.endswith('_x') and col[:-2] + '_y' in allotment_2024_609.columns]

# Drop the _y version and rename _x to original name
for base in dup_cols:
    allotment_2024_609.drop(columns=[base + '_y'], inplace=True)
    allotment_2024_609.rename(columns={base + '_x': base}, inplace=True)

allotment_2024_609
```

```python
ppr_oc_cs = pd.read_excel('/content/drive/MyDrive/TNEA Seat Matrix 2025/CSE(OC).xlsx')
ppr_oc_cs
```

```python
# Define region keywords
chennai_keywords = ['chennai', 'kancheepuram', 'chengalpattu', 'thiruvallur']

# Convert College Name to lowercase and filter by region
ppr_oc_cs['Lower College Name'] = ppr_oc_cs['College Name'].str.lower()
df_chennai_region = ppr_oc_cs[ppr_oc_cs['Lower College Name'].str.contains('|'.join(chennai_keywords))]

# Sort by Public Perception Rank and get top 10
df_top10_chennai = df_chennai_region.sort_values(by='PUBLIC PERCEPTION RANK').head(100)

# Drop the helper column and display
df_top10_chennai = df_top10_chennai.drop(columns='Lower College Name')
df_top10_chennai[['College Code', 'College Name', 'Course Code', 'PUBLIC PERCEPTION RANK']]
```

```python
df_top10_chennai = df_top10_chennai.rename(columns={
    'College Code': 'COLLEGE CODE',
    'Course Code': 'BRANCH CODE',
    # Add more mappings as needed
})
```

```python
# map df_top10_chennai to alotment_2024_609 by college code and Course code
# Ensure columns are clean and match in both DataFrames
df_top10_chennai['COLLEGE CODE'] = df_top10_chennai['COLLEGE CODE'].astype(str).str.zfill(4)
df_top10_chennai['BRANCH CODE'] = df_top10_chennai['BRANCH CODE'].str.upper().str.strip()

allotment_2024_609['COLLEGE CODE'] = allotment_2024_609['COLLEGE CODE'].astype(str).str.zfill(4)
allotment_2024_609['BRANCH CODE'] = allotment_2024_609['BRANCH CODE'].str.upper().str.strip()

# Merge on College Code and Course Code
merged_df = pd.merge(
    allotment_2024_609,
    df_top10_chennai,
    on=['COLLEGE CODE', 'BRANCH CODE'],
    how='inner'
)

# Display result
merged_df
```
