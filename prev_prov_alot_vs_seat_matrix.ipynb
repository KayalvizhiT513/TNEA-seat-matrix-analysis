## Extract tables

### Subtask:
Read the PDF file and extract tables from each page using the installed library.


**Reasoning**:
Read the PDF file and extract tables from each page using the installed library.



```python
from tabula import read_pdf

pdf_path = '/content/GENERAL_ACADEMIC_ROUND3_2024.pdf'
tables = read_pdf(pdf_path, pages='all', multiple_tables=True)
```

## Combine tables

### Subtask:
Concatenate or merge the extracted tables from all pages into a single DataFrame.

**Reasoning**:
Concatenate the list of DataFrames into a single DataFrame using `pd.concat()`.

```python
combined_df = pd.concat(tables)
display(combined_df.head())
```

## Save to Excel

### Subtask:
Save the combined DataFrame to an Excel file.

**Reasoning**:
Save the `combined_df` to an Excel file named `combined_tables.xlsx`.

```python
combined_df.to_excel('GENERAL_ACADEMIC_ROUND3_2024.xlsx', index=False)
```
